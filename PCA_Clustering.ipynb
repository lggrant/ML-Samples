(b)	Print out the corresponding PCA loading vectors.

(c)	Calculate the variance explained by each principal component. Note the sum of them should equal to 17, why? 


```r
# Conduct principal components analysis with scaling
pca_result <- prcomp(College, scale. = TRUE)

# (a) Check the center and scale components
pca_center <- pca_result$center
pca_scale <- pca_result$scale

# (b) Print out the corresponding PCA loading vectors
pca_loadings <- pca_result$rotation

# Display the results
pca_center
pca_scale
pca_loadings
```

```r
pca_var <- pca_result$sdev^2
# Display the variance explained by each principal component
pca_var
# Sum of the variances should be equal to the number of variables (17)
sum(pca_var)
```

It equals 17 because there is 17 variables in the dataset after removing the categorial variable.

The PCA biplot visualizes the relationships among 50 randomly selected universities based on their principal components. The arrows represent key variables influencing the universities' characteristics, with longer arrows indicating stronger contributions. Universities can use this information for benchmarking, strategic planning, and improving competitiveness in admissions and funding decisions.

```r
# Calculate PVE
pve <- pca_var / sum(pca_var)
cumulative_pve <- cumsum(pve)
# Plot PVE and cumulative PVE
par(mfrow = c(1, 2)) # for side-by-side plots
# Plot PVE for each component
plot(pve, xlab = "Principal Component", ylab = "Proportion of Variance Explained", type = "b", main = "PVE by Principal Components")
# Plot cumulative PVE
plot(cumulative_pve, xlab = "Principal Component", ylab = "Cumulative PVE", type = "b", main = "Cumulative PVE")
```

(10 points) With function kmeans() to perform K-means clustering with K = 2 and plot your clustering result. Set nstart=20. Report total within-cluster sum of squares 

```r
set.seed(2) 
x <- matrix(rnorm(50 * 2), ncol = 2) 
x[1:25, 1] <- x[1:25, 1] + 3 
x[1:25, 2] <- x[1:25, 2] - 4 

km.out <- kmeans(x, 2, nstart = 20)
km.out
km.out$cluster
```

The total within-cluster sum of squares is **72.8%**.

```r
par(mfrow = c(1,2))
plot(x, col = (km.out$cluster + 1), main = 'K-Means Clustering Results with K = 2', xlab = '', ylab = '', pch = 20, cex = 2)
```

V. (20 Points) Use hclust() function to implement hierarchical clustering. Use dist() function to compute the inter-observation Euclidean distance matrix. a) Perform clustering observations using complete linkage. b) Perform clustering observations using average linkage. c) Perform clustering observations using single linkage. d) Plot the dendrograms of a), b) and c)

```r
# Compute distance matrix
dist_matrix <- dist(WM_comparison, method = "euclidean")

# Perform hierarchical clustering with different linkage methods
hc_complete <- hclust(dist_matrix, method = "complete")
hc_average <- hclust(dist_matrix, method = "average")
hc_single <- hclust(dist_matrix, method = "single")

# Plot dendrograms
par(mfrow = c(1, 3))
plot(hc_complete, main = "Complete Linkage", xlab = "", sub = "", cex = 0.9)
plot(hc_average, main = "Average Linkage", xlab = "", sub = "", cex = 0.9)
plot(hc_single, main = "Single Linkage", xlab = "", sub = "", cex = 0.9)
```

VI. (15 points) Use of function cutree() a) Use function cuttree() to identify the observations to 2 groups by complete linkage, average linkage, and single linkage b) Use function cuttree() to identify the observations to 4 groups by complete linkage, average linkage, and single linkage c) Compare a) and b) and interpret your results.

```r
#(a)Cut into 2 clusters
cut_complete_2 <- cutree(hc_complete, k = 2)
cut_average_2 <- cutree(hc_average, k = 2)
cut_single_2 <- cutree(hc_single, k = 2)

# (b) Cut into 4 clusters
cut_complete_4 <- cutree(hc_complete, k = 4)
cut_average_4 <- cutree(hc_average, k = 4)
cut_single_4 <- cutree(hc_single, k = 4)

# Compare the cluster assignments
comparison_2_clusters <- table(cut_complete_2, cut_average_2, cut_single_2)
comparison_4_clusters <- table(cut_complete_4, cut_average_4, cut_single_4)

print(comparison_2_clusters)
print(comparison_4_clusters)
```

For the 2 cluster solution, 50 universities are assigned to cluster 1 by all three methods (complete, average, and single). 1 university is assigned to cluster 2 by all three methods and there is perfect agreement among all three clustering methods for the 2-cluster solution.

For the 4 cluster solution, 42 universities are assigned to cluster 1 by all three methods.
Six universities are assigned to cluster 2 by both complete and average linkage. One university is assigned to cluster 3 by complete linkage, one university is assigned to cluster 2 by average linkage, and one university is assigned to cluster 2 by single linkage. One university is assigned to cluster 3 by complete linkage, one university to cluster 3 by average linkage, and one university to cluster 2 by single linkage. One university is assigned to cluster 4 by all three methods.

Overall,the 2-cluster solution, there's perfect agreement across all methods. For the 4-cluster solution, there's high agreement for cluster 1 and cluster 4, but some disagreement for clusters 2 and 3. Single linkage tends to group more universities into cluster 1 compared to the other methods in the 4-cluster solution.
